# Import required libraries

```python
import numpy as np
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
```

# Load the dataset

```python
pokemonData = pd.read_csv('Pokemon.csv')
```

# Generate the visualizations

```python
# Create a Bar Chart

# Extract the desired column
type1 = pokemonData['Type 1']

# Get the number of Pokemons for each type (type 1)
typeCount = Counter(type1)

# Get the required parameters for bar plot 
x = typeCount.keys() # Type names
height = typeCount.values() # Type counts

# Draw the bar plot
plt.figure(figsize = (20,10))

     # Set colors for each bar
Colors = ["#BBBDAC","#AD5CA2", "#F8E64A", "#F95643", "#F9AEFE", "#A35442",
          "#FB61B8", "#CDBD78", "#7673DA", "#66EBFF", "#8B76FF","#8E6854", 
          "#C3C1D6", "#75A4F9", "#8ED758", "#F6CA42","#53AFFE", "#C3D228"] 
plt.bar(x, height, color = Colors)

    # Set title and axes labels
plt.xlabel('Type name')
plt.ylabel('Type count')
plt.title('Number of Pokemons for each type')
```

The first visualization I create is a Bar Chart, which shows the number of Pokemons for each type (type 1). I once thought about Line Chart but decided not to use it. Because Line Chart is more suitable for the ordinal variables (ordered in time series or measurement levels), helping people view a general trend or change. For example, if I want to present the annual salary from 2008 to 2018, I would choose Line Chart since we can view not only salary in each year but also general trend/change of salary level. Back to our case, the Type name is not ordinal, and we don't need to find a general trend as well. So I used the Bar Chart instead of Line Chart.

Advantage: Bar Chart can clearly show comparison among types, so we can easily get information like which type is the most prevalent or the rarest. It also presents all categories (types) in a simple, straightforward way, easily understood by viewers.

Disadvantage: The Bar Chart doesn't generate labels for specific count of each type on top of the bars. Then we can't directly get those numerical values from it. So Bar Chart sometimes may need additional information as complement. 

```python
# Create a Pie Chart

plt.figure(figsize = (10, 10))

# The height is from the previous part, which represents the number of Pokemons for each type
# The parameter autopct can show us the corresponding percent value of each type's count
# The parameter labels is assigned equal to x, which represents the list of type names
# Also different colors are used to identify types
plt.pie(height,labels = x, colors = Colors, autopct = '%1.1f%%')

# Set the plot title
plt.title('Number of Pokemons for each type in percent value')
```

The second visualization I create is a Pie Chart, which shows the number of Pokemons for each type (type 1) in percent value. Since we know the dataset totally contains 800 unique records, it is simple to calculate the actual counts (800 * corresponding percent value of each type). I once thought about drawing a Table for visualization. But I realized that presenting data via Table may be a little bit tedious for viewers. Though the Table has clear structure, it will contribute less to impress viewers and enhance viewers' memory. So I chose the Pie Chart, which more interests me, as my second visualization.

Advantage: The Pie Chart also shows clear comparison among categories, based on areas of each slice. It also can save space since other visualizations like Bar Chart and Table need wide space to present all category names.  

Disadvantage: The Pie Chart shows the actual count in the form of percent value, might be inconvenient for people with less mathematical background. And it may become messy and less effective if too many slices are presented. Also the Pie Chart can generate lie factor and occlusion due to impact of visual angle, as we discussed in class (mentioned in slide "Rules of Thumb & Design" pg.30 and slide "Marks & Channels" pg.30-32).   



For me, I prefer the first visualization (Bar Chart). Because we always want to minimize the lie factor and avoid misinterpretation when present data to viewers. Also Bar Chart is more straightforward, widespreadly accepted and used for visualizing count of categorical variable.
